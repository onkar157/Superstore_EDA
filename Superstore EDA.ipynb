```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
plt.style.use("dark_background")
```

```python
df = pd.read_csv("C:/Users/Onkar/Downloads/Superstore.csv",  encoding ='latin1' )
```

```python
#Add Profit margin clm 

df['Profit margin'] = df['Profit'] / df['Sales']*100
df.columns
```

```python
df[['Sales', 'Quantity','Discount','Profit','Profit margin']].describe().T
```

```python
#Heatmap
corr_matrix = df[['Sales', 'Quantity','Discount','Profit','Profit margin']].corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot= True, cmap='viridis')
plt.title("Correlation matrix", fontsize = 18)
plt.show()
```

```python
#Top 10 customer By Sales
top_cust = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize = (12,6))
top_cust.plot(kind= 'bar', color='teal')
plt.title("Top 10 Customer By Sales", fontsize= 20)
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.show()
```

```python
#Top 10 Best Selling Products
top_products = df.groupby('Product ID')['Sales'].sum().sort_values(ascending= False).head(10)
plt.figure(figsize = (12,6))
top_products.plot(kind='bar', color='teal')
plt.title("Top 10 Best Selling Products", fontsize=20)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.show()

```

```python
# Top 10 Most Profitable Cities
Profitable_cities = df.groupby('City')['Profit'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
Profitable_cities.plot(kind='bar', color='teal')
plt.title("Top 10 Profitable Cities", fontsize=20)
plt.xlabel("City")
plt.ylabel("Profit")
plt.show()
```

```python
#Grouping by city and calculating sum of sales
top_city = df.groupby(['City'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)
plt.figure(figsize=(12,6))
bars = plt.bar(top_city['City'], top_city['Sales'], color='teal')
plt.title("Top 10 cities by total sales", fontsize=20)
plt.xlabel("City", fontsize= 12)
plt.ylabel("Total sales", fontsize= 12)
plt.xticks(rotation=45, ha='center', fontsize= 14)
plt.show()
```

```python
segmentwise_quantity = df.groupby('Segment')['Quantity'].sum()
print("Total quantity sold by each segment :-")
print(segmentwise_quantity)
print('--------------------------------------------------------')

segwise_sale = df.groupby('Segment')['Sales'].sum()
print("Total sales by segment :- ")
print(segwise_sale)
print('--------------------------------------------------------')

segwise_profit = df.groupby('Segment')['Profit'].sum()
print("Total profit by segment :- ")
print(segwise_profit)

```

```python
#sales and profit distribution by segment
color = ['crimson', 'blueviolet', 'teal']
segwisee_sale = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
segwisee_profit = df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
fig,axes = plt.subplots(1,2, figsize=(18,12))
axes[0].pie(segwisee_sale, labels=segwisee_sale.index, autopct='%1.2f', colors=color)
axes[0].set_title('Segment Wise Sales Distribution', fontsize=14)
axes[1].pie(segwisee_profit, labels=segwisee_profit.index, autopct='%1.2f', colors=color)
axes[1].set_title("Segment Wise Profit Distribution", fontsize=14)
plt.show()

```

```python
#Segment wise profit margin

seg_profit_margin = df.groupby('Segment')['Profit margin'].mean().sort_values(ascending=False)
plt.figure(figsize=(30,20))
bars = seg_profit_margin.plot(kind='bar', color='teal')
plt.xlabel('segment', fontsize=50)
plt.ylabel('profit margin (%)', fontsize=50)
plt.xticks(fontsize=40)
plt.yticks(fontsize=40)
plt.show()
```

```python
#sub-category wise sales and its percentage 
subwise_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print("Sub-category wise sales :- ")
print()
print(subwise_sales)
print("-----------------------------------------------------------")

total_sales = df['Sales'].sum()
subcatwise_sales = (df.groupby('Sub-Category')['Sales'].sum())/total_sales * 100
plt.figure(figsize=(12,8))
ax = subcatwise_sales.sort_values(ascending=True).plot(kind='barh',color='teal')
plt.title("Sub-Category Wise Sales Percentage", fontsize=30)
plt.xlabel("Percentage of total sales", fontsize=15)
plt.ylabel("Sub-Category", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#Total sales by region
plt.figure(figsize=(12,6))
sr = df.groupby('Region', as_index=False)['Sales'].sum()
bars=plt.bar(sr['Region'], sr['Sales'], color='teal')
plt.xlabel("Region", fontsize=15)
plt.ylabel("Sales", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
# plt.show()

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            f'{bar.get_height():.2f}', ha='center', va='bottom', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax =sns.countplot(x='Ship Mode', data=df)
plt.title('Count of orders shipped by each mode', fontsize=20)
plt.xlabel("Ship Mode", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

for i in ax.patches:
    ax.annotate(f'{i.get_height()}', (i.get_x() + i.get_width()/2., i.get_height()),
                ha='center', va='center', fontsize=15)
    
plt.show()
```

```python
plt.figure(figsize=(12,6))
ax= df['Customer Name'].value_counts().head(10).plot(kind='bar', color='teal')
plt.title("Frequency of purchase for each customer", fontsize=20)
plt.xlabel('Customer Name', fontsize=15)
plt.ylabel('Frequency', fontsize=15)
plt.xticks(rotation=45,fontsize=12, ha='right')
plt.grid(axis='y', linestyle='-', alpha=0.2)

for i in ax.patches:
    ax.annotate(f'{i.get_height()}', (i.get_x() + i.get_width()/2., i.get_height()),
                ha='center', va='center', fontsize=15)
plt.show()

```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['YearMonth'] = df['Order Date'].dt.to_period('M')

month_data = df.groupby('YearMonth').agg({'Sales':'sum', 'Profit':'sum'})
plt.figure(figsize=(12,8))
plt.plot(month_data.index.astype(str),month_data['Sales'], marker='o')
plt.plot(month_data.index.astype(str),month_data['Profit'],marker='+')
plt.title("Monthly sales and profit trends", fontsize=20)
plt.xlabel("Year-Month", fontsize=15)
plt.ylabel("Amount", fontsize=15)
plt.xticks(rotation=90,fontsize=12)
plt.yticks(fontsize=15)
plt.legend(['sales','profit'])
plt.grid(axis='y', linestyle='-', alpha=0.2)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(12,8))
ax= sns.barplot(x='Region', y='Profit margin', data=df,ci=None)
plt.title("Region wise profit margin", fontsize=20)
plt.xlabel('Region', fontsize=20)
plt.ylabel('Profit margin', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='-', alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax= sns.violinplot(x='Category', y='Profit margin', data=df)
plt.title("Profit margin distribution by category", fontsize=20)
plt.xlabel('Category', fontsize=20)
plt.ylabel('Profit margin', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='-', alpha=0.2)
plt.tight_layout()
plt.show()
```

```python
sub_catwise_sale = df.groupby('Sub-Category')['Sales'].sum()
avg_discount = df.groupby('Sub-Category')['Discount'].sum()

plt.figure(figsize=(16,12))
plt.subplot(2,1,1)
sub_catwise_sale.plot(kind='line',marker='o')
plt.title('Sub category wise sales', fontsize=20)
# plt.xlabel('Sub-category', fontsize=15)
plt.ylabel('Sales', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='-', alpha=0.2)

plt.subplot(2,1,2)
avg_discount.plot(kind='line',marker='+')
plt.title('Avg Discount per Sub category', fontsize=20)
plt.xlabel('Sub category', fontsize=15)
plt.ylabel('Avg Discount', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(axis='y', linestyle='-', alpha=0.2)
plt.show()
```
